```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
#Ingreso total por tienda
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingreso total tienda 1:", ingreso_tienda1)
print("Ingreso total tienda 2:", ingreso_tienda2)
print("Ingreso total tienda 3:", ingreso_tienda3)
print("Ingreso total tienda 4:", ingreso_tienda4)

```

```python
#categorias mas populares

ventas_t1 = tienda.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_t2 = tienda2.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_t3 = tienda3.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_t4 = tienda4.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)

print("Tienda 1")
print(ventas_t1, "\n")

print("Tienda 2")
print(ventas_t2, "\n")

print("Tienda 3")
print(ventas_t3, "\n")

print("Tienda 4")
print(ventas_t4, "\n")
```

```python
# calificación promedio por tienda
calif_t1 = tienda["Calificación"].mean()
calif_t2 = tienda2["Calificación"].mean()
calif_t3 = tienda3["Calificación"].mean()
calif_t4 = tienda4["Calificación"].mean()

print("Tienda 1:", round(calif_t1, 2))
print("Tienda 2:", round(calif_t2, 2))
print("Tienda 3:", round(calif_t3, 2))
print("Tienda 4:", round(calif_t4, 2))
```

```python
# producto más y menos vendido en cada tienda

def productos_top_bottom(df, nombre_tienda):
    conteo = df["Producto"].value_counts()

    mas_vendido = conteo.idxmax()
    menos_vendido = conteo.idxmin()

    print(f"{nombre_tienda}: ")
    print("Producto más vendido:", mas_vendido, ":", conteo.max(), "ventas")
    print("Producto menos vendido:", menos_vendido, ":", conteo.min(), "ventas\n")

productos_top_bottom(tienda, "Tienda 1")
productos_top_bottom(tienda2, "Tienda 2")
productos_top_bottom(tienda3, "Tienda 3")
productos_top_bottom(tienda4, "Tienda 4")
```

```python
# Costo de envío promedio por tienda
costo_t1 = tienda["Costo de envío"].mean()
costo_t2 = tienda2["Costo de envío"].mean()
costo_t3 = tienda3["Costo de envío"].mean()
costo_t4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio")
print("Tienda 1:", round(costo_t1, 2))
print("Tienda 2:", round(costo_t2, 2))
print("Tienda 3:", round(costo_t3, 2))
print("Tienda 4:", round(costo_t4, 2))
```

```python
import matplotlib.pyplot as plt

ingresos = [
    ingreso_tienda1,ingreso_tienda2,ingreso_tienda3,ingreso_tienda4
]

labels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(labels, ingresos)
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos")
plt.show()
```

```python
ventas_cat_t1 = ventas_t1
ventas_cat_t1.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 1")
plt.ylabel("Cantidad de ventas")
plt.xlabel("Categoría del Producto")
plt.show()

ventas_cat_t2 = ventas_t2
ventas_cat_t2.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 2")
plt.ylabel("Cantidad de ventas")
plt.xlabel("Categoría del Producto")
plt.show()

ventas_cat_t3 = ventas_t3
ventas_cat_t3.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 3")
plt.ylabel("Cantidad de ventas")
plt.xlabel("Categoría del Producto")
plt.show()

ventas_cat_t4 = ventas_t4
ventas_cat_t4.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 4")
plt.ylabel("Cantidad de ventas")
plt.xlabel("Categoría del Producto")
plt.show()
```

```python
calificaciones = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.plot(labels, calificaciones, marker="o")
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Promedio de calificación")
plt.show()
```
